```python
from aatm.data_models import ExtractionTask
from aatm.extractors import GeminiExtractor


text = """
Patient: Maria Aparecida Santos
Age: 67 years
Sex: Female
Date: 2026-04-23

S — Subjective

Chief Complaint:
“I've been having a headache for the past 3 days.”

History of Present Illness:
Patient reports a gradual-onset headache over the past 3 days, described as a bilateral, pressing pain located in the frontal and occipital regions. Pain intensity is 6/10, persistent throughout the day, slightly worse in the evening. No pulsatile quality. No associated photophobia, phonophobia, nausea, or vomiting. Denies visual disturbances, focal weakness, speech difficulty, or loss of consciousness.

She reports increased stress and poor sleep over the past week. Has been using dipyrone intermittently with partial relief.

Past Medical History:

Type 2 Diabetes Mellitus (15 years)
Hypertension (20 years)
Chronic Kidney Disease stage 3
Dyslipidemia
Osteoarthritis (knees)

Medications:

Metformin 850 mg BID
Insulin glargine 20U nightly
Losartan 50 mg BID
Amlodipine 5 mg daily
Atorvastatin 20 mg nightly
Paracetamol PRN

Allergies:
No known drug allergies

Social History:

Retired
Lives with husband
Denies tobacco, alcohol, or illicit drug use

Family History:

Mother: stroke
Father: myocardial infarction
O — Objective

Vital Signs:

BP: 158/92 mmHg
HR: 82 bpm
RR: 16/min
Temp: 36.5°C
SpO₂: 98% on room air

General:
Alert, oriented, no acute distress

Neurological Exam:

Cranial nerves II-XII intact
Motor strength 5/5 in all extremities
Sensation intact
No focal deficits
Normal gait

Head and Neck:

Mild tenderness in cervical paraspinal muscles
No temporal artery tenderness

Cardiovascular:
Regular rhythm, no murmurs

Fundoscopy:
No papilledema

A — Assessment
Tension-type headache — most likely given bilateral, non-pulsatile pain, stress trigger, and lack of red flags
Uncontrolled hypertension — may be contributing to headache
Type 2 Diabetes Mellitus — chronic, stable
Chronic Kidney Disease stage 3 — stable
Multimorbidity with polypharmacy — requires careful medication management
P — Plan

Headache Management:

Continue paracetamol PRN (max 3g/day)
Avoid NSAIDs due to CKD
Recommend non-pharmacological measures: hydration, sleep hygiene, stress reduction

Hypertension:

Reinforce adherence
Increase amlodipine to 10 mg daily
Home BP monitoring

Diabetes:

Maintain current regimen
Schedule HbA1c check

Investigations:

Basic labs: CBC, electrolytes, renal function
HbA1c
Lipid panel

Follow-up:

Return in 2 weeks for reassessment
Earlier if worsening headache or red flag symptoms (neurological deficits, severe sudden headache, vision changes)

Patient Education:

Warning signs discussed
Lifestyle counseling (diet, sleep, stress)
"""
```

```python
from aatm.omop.condition_occurrence import ListOfConditionOccurrenceExtractionModel


prompt_template = """Your task is to extract data related to condition occurrence in the OMOP common data model. Only extract information that is present in the context given. Do not use your world knowledge to fill in information that is not explicitly stated in the context provided. Return a JSON string following the required format.

# Text
{text}

# JSON schema
{json_schema}
"""

data_model = (
    "condition_occurrence_list"  # or ListOfConditionOccurrenceExtractionModel object
)

# You may define one or multiple tasks
task = ExtractionTask(
    prompt_template=prompt_template,
    texts=[text],
    prompt_args={
        "json_schema": ListOfConditionOccurrenceExtractionModel.model_json_schema()
    },
    data_model=data_model,
)

extractor = GeminiExtractor(model_id="gemini-2.5-flash")
```

```python
# extractors expect a list of tasks
results = extractor([task])
```
