```python
import chromadb
import dotenv

from aatm.retrievers import ChromaDBRetriever
from aatm.embedding_functions import GoogleEmbeddingFunction
from aatm.translators import GeminiTranslator
from aatm.selectors import FirstResultSelector

dotenv.load_dotenv()

client = chromadb.PersistentClient()
translator = GeminiTranslator(model='gemini-2.5-flash')
retriever = ChromaDBRetriever(
    client=client,
    collection_name="expressions",
    embedding_function=GoogleEmbeddingFunction(model="gemini-embedding-001"),
)
selector = FirstResultSelector()
```

```python
results = "COVID-19 Prevention in Schools" | retriever 
for idx, result in enumerate(results.results[0]):
    print(f"{idx+1} - {result.expression}")
    print(f"STD concept name: {result.std_concept_name}")
    print(f"STD vocabulary code: {result.std_vocabulary_code}")
    print(f"STD concept id: {result.std_concept_id}")
    print(f"STD vocabulary id: {result.std_vocabulary_id.value}")
    print()
```
