```python
import pandas as pd
from pathlib import Path

filepath = Path(r"data\TUSS\Tabela_22_Procedimentos_e_eventos_em_saúde.csv")
df = pd.read_csv(filepath, sep=";", encoding="utf-8")
df
```

```python
relevant_cols = [
                "source_code",
                "source_concept_id",
                "source_vocabulary_id",
                "source_code_description",
                "valid_start_date",
                "valid_end_date",
                "invalid_reason",
            ]

df = df.rename(columns={
    "Código do Termo": "source_code",
    "Termo": "source_code_description",
})
df['source_concept_id'] = df['source_code']
df['source_vocabulary_id'] = "TUSS"
df['valid_start_date'] = df["Data de início de vigência"]
df['valid_end_date'] = df["Data de fim de vigência"]
df['invalid_reason'] = None
df[relevant_cols].to_csv('omop_tuss_procedures_source_to_concept_map.csv')
```
