```python
%cd c:\Users\almei\Documents\GitHub\precision-data
from pathlib import Path
from aatm.aio.translators import EmptyTranslator
from aatm.selectors import GeminiLLMModels
from aatm.terminology_mapper import TerminologyMapper
from aatm.retrievers import CHROMADB_RETRIEVER_MODEL_REGISTRY as model_registry
from aatm.retrievers import load_chromadb_retriever
from aatm.aio.selectors import AsyncGeminiLLMSelector, AsyncOpenAILLMSelector

for model_name in [
    # "text-embedding-3-large", 
    "text-embedding-3-small", 
    "qwen3-06B", 
    "qwen3-4B",
    "embeddinggemma-300M",
    #   "gemini-embedding-001",
    ]:
    for selector_name in [
                        # "gpt-5.2", 
                        # "gpt-5", 
                        # "gpt-5-nano", 
                        # "gpt-5-mini",
                        # GeminiLLMModels.GEMINI_2_5_FLASH_LITE.value,
                        GeminiLLMModels.GEMINI_2_5_FLASH.value,
                        # GeminiLLMModels.GEMINI_2_5_PRO.value,
                        ]:
        
        print(f"Processing {model_name} with {selector_name}")

        translator = EmptyTranslator()
        retriever = load_chromadb_retriever(model_name)
        selector = AsyncGeminiLLMSelector(selector_name)

        tm = TerminologyMapper(
            translator=translator,
            retriever=retriever,
            selector=selector, 
            batch_size=10,
            rate_limit=model_registry[model_name].get("rate_limit"),
            output_dir=Path(model_registry[model_name]["output_path"])/"with-llm-selector"/selector_name,
            
        )

        mapped_concepts_df = await tm.amap(
            'vocabularies/SOURCE_TO_CONCEPT_MAP.csv',
        )
```
